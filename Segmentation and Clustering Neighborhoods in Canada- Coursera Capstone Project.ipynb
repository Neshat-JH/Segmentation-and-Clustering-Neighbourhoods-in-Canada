```python
#Here, I'm going to use beautifulsoup to get the data and the table from wiki
from bs4 import BeautifulSoup
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)

from IPython.display import display_html
table = str(soup.table)
display_html(table,raw=True)
```

```python
# Now I'm converting the html to pandas dataframe for ease of processing
dfs = pd.read_html(table)
df=dfs[0]

#checking five rows to make sure all is good!
df.head()
```

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

## 2nd Section - For Peer Grading Purposes!

In this section we need to get the latitude and the longitude coordinates of each neighborhood.

```python
#importing the latitude and longitude file provided on the submission page
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
#Now I'm going to merge the two tables so we have the latitude and longiture of each neighbourhood

lat_lon.rename(columns={'Postal Code':'Postcode'},inplace=True)
df3 = pd.merge(df2,lat_lon,on='Postcode')
df3.head()
```

## 3rd Section- For Peer Grading Purposes!

In this section, I will explore and apply Clustering on Neighbourhoods of Toronto.

```python
#So far the data we were wranggling include multiple Broughts in Canada. 
#Now we want to focus on Toronto. Thus, we would select the rows that have Toronto included in their Brought.

df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

## Note: If the maps weren't showing in Github notebook please check the README. I will provide the pictures there.

```python
#Here we are going to use visualization and Folium to display the Boroughs.
# The same syntax of data Clustering for New York City in the previous labs were used.

map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
#Here I'm going to using KMeans Clustering method to group Toronto's Boroughs.
k=5
toronto_clustering = df4.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster-Labels', kmeans.labels_)
df4
```

```python
#Again we are going to create the map but this time with the clusters.
#please check README. if no map was displayed in Github notebook.

# creating the map using the syntax presented in the previous lab (NYC example)
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster-Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
